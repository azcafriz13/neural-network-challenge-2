## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
display(y_df.head())
display(y_df['Attrition'].unique().tolist())
y_df['Department'].unique().tolist()
```

```python
# Create a list of at least 10 column names to use as X data
X_Columns = [
 'BusinessTravel',
 'Education',
 'HourlyRate',
 'JobRole',
 'JobSatisfaction',
 'PercentSalaryHike',
 'PerformanceRating',
 'WorkLifeBalance',
 'YearsAtCompany',
 'YearsInCurrentRole'
  ]

# Create X_df using your selected columns
X_df = attrition_df[X_Columns]
display(X_df.head())

# Show the data types for X_df
X_df.dtypes
```

```python
# Review unique values of new select ten columns
X_df.nunique()
```

```python
# Review the different values in BusinessTravel and JobRole
display(X_df["BusinessTravel"].value_counts())
display(X_df["JobRole"].value_counts())
```

```python
# Split the data into training and testing sets
y_attrition = y_df[['Attrition']]
y_department = y_df[['Department']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_attrition_train, y_attrition_test, y_department_train, y_department_test,= train_test_split(X_df, y_attrition, y_department)
```

## Preprocess X data

```python
# Convert 'BusinessTravel' and 'JobRole' to numeric data types.
# Continue Data Discovery: Review the unique values in BusinessTravel and JobRole
display(X_train['BusinessTravel'].unique().tolist())
display(X_train['JobRole'].unique().tolist())
```

```python
# Convert 'BusinessTravel' and 'JobRole' with onehotencoder as data is not ordinal.
# Preprocess X_train
from sklearn.preprocessing import OneHotEncoder

# Preprocess "BusinessTravel" column (one-hot encoding)
BusinessTravel_encoder = OneHotEncoder(sparse_output=False)
BusinessTravel_encoded = BusinessTravel_encoder.fit_transform(X_train[['BusinessTravel']])
BusinessTravel_columns = BusinessTravel_encoder.get_feature_names_out(['BusinessTravel'])
df_BusinessTravel_encoded = pd.DataFrame(BusinessTravel_encoded, columns=BusinessTravel_columns)

display(df_BusinessTravel_encoded.head())

# Preprocess "JobRole" column (one-hot encoding)
JobRole_encoder = OneHotEncoder(sparse_output=False)
JobRole_encoded = JobRole_encoder.fit_transform(X_train[['JobRole']])
JobRole_columns = JobRole_encoder.get_feature_names_out(['JobRole'])
df_JobRole_encoded = pd.DataFrame(JobRole_encoded, columns=JobRole_columns)

display(df_JobRole_encoded.head())
```

```python
# Find the percentage of null values in each column
df_BusinessTravel_encoded.isna().sum()/len(df_BusinessTravel_encoded)
```

```python
# Concatenate the encoded columns to the original DataFrame

X_train_processed = pd.concat([X_train.reset_index(drop=True), df_BusinessTravel_encoded.reset_index(drop=True)], axis=1)
X_train_processed = pd.concat([X_train_processed, df_JobRole_encoded.reset_index(drop=True)], axis=1)

# Drop the original "BusinessTravel" and "Jobrole" columns
X_train_processed = X_train_processed.drop(['BusinessTravel', 'JobRole'], axis=1)

display(X_train_processed.columns.tolist())

X_train_processed.tail()
```

```python
# Encode Testing Data (X_test)

from sklearn.preprocessing import OneHotEncoder

# Preprocess "BusinessTravel" column (one-hot encoding)
BusinessTravel_encoder = OneHotEncoder(sparse_output=False)
BusinessTravel_encoded = BusinessTravel_encoder.fit_transform(X_test[['BusinessTravel']])
BusinessTravel_columns = BusinessTravel_encoder.get_feature_names_out(['BusinessTravel'])
df_BusinessTravel_encoded = pd.DataFrame(BusinessTravel_encoded, columns=BusinessTravel_columns)

# Preprocess "JobRole" column (one-hot encoding)
JobRole_encoder = OneHotEncoder(sparse_output=False)
JobRole_encoded = JobRole_encoder.fit_transform(X_test[['JobRole']])
JobRole_columns = JobRole_encoder.get_feature_names_out(['JobRole'])
df_JobRole_encoded = pd.DataFrame(JobRole_encoded, columns=JobRole_columns)

# Concatenate the encoded columns to the original DataFrame
X_test_processed = pd.concat([X_test.reset_index(drop=True), df_BusinessTravel_encoded.reset_index(drop=True)], axis=1)
X_test_processed = pd.concat([X_test_processed, df_JobRole_encoded.reset_index(drop=True)], axis=1)

# Drop the original "BusinessTravel" and "Jobrole" columns
X_test_processed = X_test_processed.drop(['BusinessTravel', 'JobRole'], axis=1)

display(X_test_processed.columns.tolist())
X_test_processed.tail()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train_processed)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train_processed)
X_test_scaled = X_scaler.transform(X_test_processed)

display(X_train_scaled[:5])
display(X_test_scaled[:5])
```

```python
# Create a OneHotEncoder for the Department column

from sklearn.preprocessing import OneHotEncoder

Department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_department_train_encoded = Department_encoder.fit_transform(y_department_train[['Department']])
display(y_department_train_encoded)
display (len(y_department_train_encoded))

# Create three new variables by applying the encoder
# to the training and testing data
y_department_test_encoded = Department_encoder.fit_transform(y_department_test[['Department']])
display(y_department_test_encoded)
display(len(y_department_test_encoded))
```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import OneHotEncoder

Attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
y_attrition_train_encoded = Attrition_encoder.fit_transform(y_attrition_train[['Attrition']])
display(y_attrition_train_encoded)
display (len(y_attrition_train_encoded))

# Create two new variables by applying the encoder
# to the training and testing data
y_attrition_test_encoded = Attrition_encoder.fit_transform(y_attrition_test[['Attrition']])
display(y_attrition_test_encoded[:5])
display(len(y_attrition_test_encoded))
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
dimensions = X_train_scaled.shape
rows, columns = dimensions
print("Columns:", columns)

# Create the input layer
input_layer = layers.Input(shape=(X_train_scaled.shape[1],), name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)

dense_shared = layers.Dense(64, activation='relu')(shared_layer2)
```

```python
# Create a branch for Department with a hidden layer and an output layer

# Create the hidden layer
department_dense = layers.Dense(64, activation='relu')(dense_shared)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_dense)
```

```python
# Create a branch for Attrition with a hidden layer and an output layer

# Create the hidden layer
attrition_dense = layers.Dense(64, activation='relu')(dense_shared)

# Create the output layer
attrition_output = layers.Dense(2, activation='sigmoid', name='attrition_output')(attrition_dense)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'binary_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
# Fit the model
model.fit(
    X_train_scaled,
    {'department_output': y_department_train_encoded, 'attrition_output': y_attrition_train_encoded},
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'department_output': y_department_test_encoded, 'attrition_output': y_attrition_test_encoded})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department Predictions Accuracy: {test_results[2]}")
print(f"Attrition Predictions Accuracy: {test_results[1]}")
```
